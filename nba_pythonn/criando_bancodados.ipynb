```python
boxscore_players = boxscore_players.drop(columns = ['Unnamed: 0','slugLeague','numberGameTeamSeason','nameTeam','idTeam','isB2B', 'isB2BFirst','isB2BSecond','slugMatchup','countDaysRestTeam','countDaysNextGameTeam','slugTeamWinner',
'slugTeamLoser', 'countDaysRestPlayer', 'countDaysNextGamePlayer', 'isWin', 'pctFG', 'pctFT', 'hasVideo',
'urlTeamSeasonLogo', 'urlPlayerStats', 'urlPlayerThumbnail', 'urlPlayerHeadshot', 'urlPlayerActionPhoto', 'urlPlayerPhoto',
'fpts','pctFG3', 'pctFG2'])
```

```python
boxscore_players = boxscore_players.reindex(columns = ['yearSeason', 'slugSeason', 'typeSeason', 'dateGame', 'idGame','idPlayer', 
'namePlayer', 'slugTeam', 'numberGamePlayerSeason','locationGame', 'slugOpponent', 'outcomeGame', 'minutes', 'pts', 'ast', 'treb',
'stl', 'blk', 'tov', 'pf', 'fgm', 'fga', 'ftm', 'fta', 'fg2m','fg2a', 'fg3m', 'fg3a', 'oreb', 'dreb', 'plusminus'])
```

```python
#checando valores vazios
na_cols = [col for col in boxscore_players if (boxscore_players[col].isna().sum()>0)]
boxscore_players[na_cols].isna().mean().sort_values(ascending =False).mul(100)
```

```python
boxscore_players = boxscore_players.dropna(subset=['outcomeGame'])
```

```python
conn = sqlite3.connect('NBAdata.db')
```

```python
boxscore_players.to_sql(name='PlayersBoxscore', con = conn)
```

```python
# lendo a base de dados
df_read = pd.read_sql('SELECT * FROM PlayersBoxscore', con=conn)
```
